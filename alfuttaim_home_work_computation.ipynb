```
%python
import time
import datetime, pytz

def time_to_utc(naive, timezone="Asia/Dubai"):
    local = pytz.timezone(timezone)
    local_dt = local.localize(naive, is_dst=None)
    utc_dt = local_dt.astimezone(pytz.utc)
    return utc_dt

def utc_to_time(naive, timezone="Asia/Dubai"):
    return naive.replace(tzinfo=pytz.utc).astimezone(pytz.timezone(timezone))
```

```
%python
def get_centermost_point(cluster):
    
    converted_list_frommatrix=cluster.tolist()
    centroid = (MultiPoint(converted_list_frommatrix).centroid.x, MultiPoint(converted_list_frommatrix).centroid.y)
  
    return centroid
```

```

%python
from sklearn.cluster import DBSCAN
from shapely.geometry import MultiPoint
kms_per_radian = 6371.0088

def compute_cluster_per_user(list_of_points):
    
    number_points=len(list_of_points)
    coords=np.asmatrix(list_of_points)
  
    epsilon = 0.6/ kms_per_radian
    if(coords.shape[0]>=5): 
        
        coordinates=np.radians(coords)
        db = DBSCAN(eps=epsilon, min_samples=4, algorithm='ball_tree', metric='haversine').fit(coordinates)
        
        cluster_labels = db.labels_
        
        unique_labels = set(cluster_labels)
        num_clusters = len(set(cluster_labels))
        
        clusters = pd.Series([coords[cluster_labels == n] for n in range(num_clusters)])
        
        min_sample=4
        
        centermost_points = []
        for cluster in clusters.iteritems():
            if len(cluster[1]) >= min_sample:
                centermost_points.append((len(cluster[1]),get_centermost_point(cluster[1])))
                 
        
        
        return centermost_points

```

```
%python

def retreive_onlyhome_hotspot(list_of_centroids):
    if(list_of_centroids!=None):
        #print(len(list_of_centroids))
        if(len(list_of_centroids)>0):
            
            hotspot=sorted(list_of_centroids,key=lambda x: x[0], reverse=True)[0]
            
            home_lat=hotspot[1][0]
            home_lon=hotspot[1][1]
        #list_of_centroids
            return (home_lat,home_lon)
```
