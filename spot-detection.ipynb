```
# Merge entertainer_geolocation_events and Members of Dubai product on user/customer id
data_merged_sg=df.join(df_user, df.customer_id ==df_user.user_id).drop(df_user.user_id)
data_merged_sg.show()
```

```
data_merged_sg.filter(data_merged_sg.lat==0).count() #
```

```
data_merged_sg.filter(data_merged_sg.lat!=0).count() #
```

```
data_merged_sg=data_merged_sg.filter(data_merged_sg.lat!=0)
data_merged_sg.show()
```

```
data_merged_sg=data_merged_sg.drop('id','session_id','created_on_azure')
data_merged_sg.show()
```

```
# filter from 1st jan 20 to 1st aug 20
data_merged_sg_filter=data_merged_sg[(data_merged_sg.datetime>='2020-01-01') & (data_merged_sg.datetime<='2020-08-01')]
data_merged_sg_filter.show()
```

```
data_merged_sg_filter=data_merged_sg_filter.withColumn("datetime_modifiedzone",from_utc_timestamp(col("datetime"),"Asia/Dubai"))
data_merged_sg_filter.show()
```

```
data_merged_sg_filter=data_merged_sg_filter.withColumn("hour",hour(data_merged_sg_filter['datetime_modifiedzone']))
data_merged_sg_filter.show()
```

```
data_merged_sg_filter=data_merged_sg_filter.withColumn("weekday",dayofweek(data_merged_sg_filter['datetime_modifiedzone']))
data_merged_sg_filter.show()
```

```
data_merged_sg_filter=data_merged_sg_filter.drop('datetime','datetime_modifiedzone')
data_merged_sg_filter.show()
```
